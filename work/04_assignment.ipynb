```python
import pandas as pd
import matplotlib as plt

%matplotlib inline
%config InlineBackend.figure_format ='retina'
```

```python
data = pd.read_csv('../data/state_facts.tsv', delimiter='\t')
```

```python
data.head()
```

```python
data['Capital_City5']=data['Capital_pop']+data['city_5_pop']
plot = data[['USPS_code','Pop_2010','Capital_City5']].plot(stacked=True,kind='bar',figsize=(15,15),color=['b','pink'])
plot.set_xticklabels(data['USPS_code']);
plot.set_ylim([0, 50000000])
plot.set_xlabel('State (s)')
plot.set_ylabel('Population (In 10 Millions)')
plot.set_title('Population 2010 vs. Current Capital + City 5 Populations')
```
